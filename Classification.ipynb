```python
'''
Q1
@author: rgonzalez
'''
import numpy as np
import math
import matplotlib.pyplot as plt

def get_distance(data1, data2):
    points = zip(data1, data2)
    diffs_squared_distance = [pow(a - b, 2) for (a, b) in points]
    return math.sqrt(sum(diffs_squared_distance))               
  
def most_common(lst):
    return max(set(lst), key=lst.count)               

def kNeighbours(data1,data2,labels,k):
    vNei = []
    for i in range(0,data2.shape[0]):
        fit = [] 
        kNei = []        
        for j in range(0,data1.shape[0]):
            fit.append((get_distance(data1[j],data2[i]),labels[j]))
        fit.sort()
        for i in range(0,k):
            kNei.append(fit[i][1])
        vNei.append(most_common(kNei))    
    return np.array(vNei)

def accuracy(label1,label2):
    hit = 0
    for i in range(0,label1.size):
        if(label1[i]==label2[i]): hit += 1
    return (hit*100)/label1.size            
       
#Read data from file
X= np.loadtxt("iris.data",delimiter = ',',usecols=range(0,4))
Y= np.loadtxt("iris.data",dtype = np.str, delimiter = ',',usecols=(4,))
train = np.loadtxt("iris_idx_train.txt",dtype=int)
test = np.loadtxt("iris_idx_test.txt",dtype=int)
X_train = X[train,:]
X_test = X[test,:]
 
Y_train = Y[train]
Y_test = Y[test]
  
#read tags from file
label_train = np.empty(Y_train.size, dtype=int)
label_test = np.empty(Y_test.size, dtype=int)
 
for i in range(0, Y_train.size):
    if (Y_train[i]=="Iris-setosa"): label_train[i]=0
    elif (Y_train[i]=="Iris-versicolor"): label_train[i]=1
    else: label_train[i]=2
     
for i in range(0, Y_test.size):
    if (Y_test[i]=="Iris-setosa"): label_test[i]=0
    elif (Y_test[i]=="Iris-versicolor"): label_test[i]=1
    else: label_test[i]=2
  
print "Original Iris Train Classification:\n" + str(label_train)
print "KNN Iris Train Classification:\n" + str(kNeighbours(X_train, X_train, label_train, 5))
print "KNN Iris Train Accuracy:\n" + str(accuracy(kNeighbours(X_train, X_train, label_train, 5), label_train)) +"%\n\n"
 
print "Original Iris Test Classification:\n" + str(label_test)
print "KNN Iris Test Classification:\n" + str(kNeighbours(X_test, X_test, label_test, 5))
print "KNN Iris Test Accuracy:\n" + str(accuracy(kNeighbours(X_test, X_test, label_test, 5), label_test)) +"%\n\n"
 
X_3D_train = np.load("3dclothing_train.npy")
label_3D_train = np.loadtxt("3dclothing_labels_train.txt",dtype=np.str,usecols=(0,))
X_3D_test = np.load("3dclothing_test.npy")
label_3D_test = np.loadtxt("3dclothing_labels_test.txt",dtype=np.str,usecols=(0,))
 
for k in range(1,10):
    if k%2 != 0:
        print "KNN 3D Clothing Train Accuracy K="+str(k)+" : " + str(accuracy(kNeighbours(X_3D_train, X_3D_train, label_3D_train, k),label_3D_train)) + "%\n"

x = []
y = []
for k in range(1,10):
    if k%2 != 0:
        y.append(accuracy(kNeighbours(X_3D_test, X_3D_test, label_3D_test, k),label_3D_test))
        x.append(k)
# aPlot = np.vstack((y,x))
# print aPlot     
# aPlot = np.vstack((x,y))        
#Plot accuracy for 3d test set  
plt.suptitle("KNN 3D Clothing Test Data Accuracy")      
plt.plot(x,y)
plt.xlabel("K")
plt.ylabel("Accuracy (%)")  
plt.show()  
```

```python
'''
Q2

@author: robert
'''
import numpy as np
from sklearn.linear_model import LogisticRegression
# from sklearn.grid_search import GridSearchCV
from sklearn.model_selection import StratifiedKFold
import matplotlib.pyplot as plt


X_3D_train = np.load("3dclothing_train.npy")
label_3D_train = np.loadtxt("3dclothing_labels_train.txt",dtype=np.str,usecols=(0,))

X_3D_test = np.load("3dclothing_test.npy")
label_3D_test = np.loadtxt("3dclothing_labels_test.txt",dtype=np.str,usecols=(0,))

l_train = []
for i in range(0,label_3D_train.size):
    if ((label_3D_train[i]=="shirt")or(label_3D_train[i]=="jeans")):
        l_train.append(i)
        
l_test = []
for i in range(0,label_3D_test.size):
    if ((label_3D_test[i]=="shirt")or(label_3D_test[i]=="jeans")):
        l_test.append(i)
        
X_3D_train = X_3D_train[l_train,:] 
label_3D_train = label_3D_train[l_train]
 
X_3D_test = X_3D_test[l_test,:] 
label_3D_test = label_3D_test[l_test]

kFold = 0 
         
skf = StratifiedKFold(n_splits=15)
for train_index,test_index in skf.split(X_3D_train, label_3D_train):
    kFold+=1
    X_kFold_train = X_3D_train[train_index,:] 
    label_kFold_train = label_3D_train[train_index]
    print 'Kfold = ',kFold
    yTr=[]
    C=[]
    for i in range(-7,7):
        LogReg = LogisticRegression(C=10**i)
        LogReg.fit(X_kFold_train, label_kFold_train)
        yTr.append(LogReg.score(X_kFold_train, label_kFold_train))
        C.append(10**i)
    cIndex = C[yTr.index(1,)-1]
    LogRegBest=LogisticRegression(C=cIndex)
    LogRegBest.fit(X_3D_test, label_3D_test)    
    print 'Log Reg accuracy', LogRegBest.score(X_3D_test, label_3D_test),'C=',cIndex
    
    plt.figure(figsize=(10,10))
    plt.subplots_adjust(hspace=0.6)
    plt.suptitle("LOGISTIC REGRESSION WITH CROSS-VALIDATION")  
    plt.subplot(8,2,kFold)   
    plt.xlabel("C")
    plt.ylabel("Score")   
    plt.semilogx(C,yTr)
    axes = plt.gca()
    axes.set_ylim([0.50,1.1])
    plt.title("kFold= " + str(kFold))
plt.show()
    
```

```python
'''
Q4

@author: robert
'''

import numpy as np
import math
from sklearn.linear_model import LogisticRegression

def varianza(data):
    sumVar = 0
    for j in range(0,data.size):
        sumVar += (data[j]-np.mean(data))**2
    var = sumVar / data.size    
    return var

def desviacion(data):
    return math.sqrt(varianza(data))


X_3D_train = np.load("3dclothing_train.npy")
label_3D_train = np.loadtxt("3dclothing_labels_train.txt",dtype=np.str,usecols=(0,))

X_3D_test = np.load("3dclothing_test.npy")
label_3D_test = np.loadtxt("3dclothing_labels_test.txt",dtype=np.str,usecols=(0,))

l_train = []
for i in range(0,label_3D_train.size):
    if ((label_3D_train[i]=="shirt")or(label_3D_train[i]=="jeans")):
        l_train.append(i)
        
l_test = []
for i in range(0,label_3D_test.size):
    if ((label_3D_test[i]=="shirt")or(label_3D_test[i]=="jeans")):
        l_test.append(i)
        
X_3D_train = X_3D_train[l_train,:] 
label_3D_train = label_3D_train[l_train]
 
X_3D_test = X_3D_test[l_test,:] 
label_3D_test = label_3D_test[l_test]

S= []
for i in range(0,X_3D_train.shape[1]):
    S.append(desviacion(X_3D_train[:,i]))
    
S_weight = []
for i in range(0,len(S)):
    if(S[i]!=0):
        if(S[i]/max(S) >= 0.2):
            S_weight.append(i)
            
           
X_Weight_test = X_3D_test[:,S_weight] 

print "The most important information is the weight of Features."
print "In this case, I've extracte deviation of Train Features to take the weight,"
print "and I've used the most representative features to make the Logistic Regression.\n"

for i in range (-7,3):
    LogReg = LogisticRegression(C=10**i)
    LogReg.fit(X_Weight_test, label_3D_test)
    print 'Accuracy:', LogReg.score(X_Weight_test, label_3D_test),"C=",10**i
```

```python
'''
Q5

@author: robert
'''
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.svm import SVC 

# Adapted from: Gael Varoquaux, Andreas Muller; "Classifier comparison"  
# http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html  
def paint_decision_functions(data, labels, clf):  
    from matplotlib.colors import ListedColormap
    import pylab  
    cm = pylab.cm.get_cmap('RdBu')
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])  
    x_min, x_max = data[:, 0].min() - .5, data[:, 0].max() + .5  
    y_min, y_max = data[:, 1].min() - .5, data[:, 1].max() + .5  
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),  
                         np.arange(y_min, y_max, 0.1))  
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])  
    Z = Z.reshape(xx.shape)  
    pylab.contourf(xx, yy, Z, cmap=cm, alpha=.8)  
    pylab.scatter(data[:, 0], data[:, 1], c=labels, cmap=cm_bright)  
    pylab.xlim(xx.min(), xx.max())  
    pylab.ylim(yy.min(), yy.max())  
    pylab.xticks(())  
    pylab.yticks(())  
    pylab.show() 

data = np.loadtxt("jain.txt",usecols=(0,1))
label = np.loadtxt("jain.txt",usecols=(2,))

data_train,data_test,label_train,label_test = train_test_split(data,label,test_size=0.5,random_state=23)

clf =  LinearSVC(C=1)
clf.fit(data_train, label_train)
print 'Linear SVM score', clf.score(data_train, label_train)
parameters = {'C':[10**i for i in range(-3, 4)]}
clf = GridSearchCV(LinearSVC(), parameters, verbose=False)
clf.fit(data_train, label_train)
print 'Linear SVM score with Grid Search', clf.score(data_test, label_test), 'with C =', clf.best_params_['C']

paint_decision_functions(data_test, label_test, clf)

# Now we will do the same with the kernelized SVM
clf = SVC(C=1)
clf.fit(data_train, label_train)
print 'RBF SVM score', clf.score(data_test, label_test)
# We can actually define several parameters we want to optimize over:
parameters = {'C':[10**i for i in range(-3, 4)], 'kernel':['linear', 'rbf']}
clf = GridSearchCV(SVC(), parameters, verbose=False)
clf.fit(data_train, label_train)
print 'RBF SVM score with Grid Search', clf.score(data_test, label_test)
# print 'RBF SVM score with Grid Search', clf.score(data_test, label_test), 'with C =', \
#     parameters.best_params_['C'], 'and kernel =', parameters.best_params_['kernel']

paint_decision_functions(data_test, label_test, clf)

print "Boundaries are more adjusted with kernelized SVM. The color of the areas"
print "shows that the indecision area is narrower in kernelized, but all the features"
print "has been corretly classified in both models."
```
